# Interactive Pretty Graph Goes BRR Demo

**Graph types:** `line`, `scatter`, `line+scatter`, `step`, `area`, `bar`, `histogram`, `bell`/`kde`

````python
from bokeh.io import output_notebook
output_notebook()

import numpy as np
from plot_framework import iplot, isweep, ioverlay, isweep_overlay
````

## Single Plot (`iplot`)
### Line

````python
x = np.linspace(0, 10, 500)
y = np.sin(x) * np.exp(-0.2 * x)

iplot(x, y, title="Damped Sine", xlabel="Time", ylabel="Amplitude", kind="line")
````

### Scatter

````python
x = np.random.randn(300)
y = 0.6 * x + 0.4 * np.random.randn(300)

iplot(x, y, title="Noisy Correlation", xlabel="x", ylabel="y", kind="scatter")
````

### Histogram

````python
samples = np.random.gamma(2, 2, size=5000)
dummy_x = np.arange(len(samples)) # Historgram doesn't use x values, but we need to pass something to make functions consistent...

iplot(dummy_x, samples, title="Gamma(2,2) Samples", xlabel="Value", ylabel="Count", kind="histogram")
````

### Bell Curve / KDE

````python
samples = np.concatenate([np.random.normal(-2, 0.8, 3000),
                          np.random.normal(2, 1.2, 4000)])
dummy_x = np.arange(len(samples)) # refer to histogram note boave

iplot(dummy_x, samples, title="Bimodal KDE", xlabel="Value", ylabel="Density", kind="bell")
````

### Log Scale Axis

````python
x = np.logspace(1, 6, 500)
y = 1e4 / x**0.5 + 0.01

iplot(x, y, title="Power Law", xlabel="Frequency", ylabel="Magnitude", kind="line", x_log=True, y_log=True)
````

## Overlay (`ioverlay`)

Click legend entries to hide/show individual traces.

````python
x = np.linspace(0, 4 * np.pi, 500)

ioverlay(
    {
        "sin(x)":       (x, np.sin(x)),
        "sin(2x)":      (x, np.sin(2 * x)),
        "sin(x) + cos(3x)": (x, np.sin(x) + np.cos(3 * x)),
    },
    title="Overlaid Waveforms",
    xlabel="x", ylabel="y",
    kind="line",
)
````

### Overlay Scatter Example

````python
datasets = {}
for mu, label in [(0, "Cluster A"), (3, "Cluster B"), (6, "Cluster C")]:
    n = 150
    datasets[label] = (np.random.normal(mu, 0.8, n),
                       np.random.normal(mu * 0.5, 1.0, n))

ioverlay(datasets, title="Clusters", xlabel="x", ylabel="y", kind="scatter")
````

## Slider Sweep (`isweep`)

Flat dict to single slider. Drag to switch datasets.

````python
x = np.linspace(0, 10, 500)

datasets = {}
for freq in [0.5, 1.0, 2.0, 4.0, 8.0]:
    datasets[f"freq = {freq}"] = (x, np.sin(2 * np.pi * freq * x / 10))

isweep(datasets, title="Frequency Sweep", xlabel="x", ylabel="y", kind="line")
````

### 1-Slider Sweep, KDE Example

Sweep across different distribution shapes.

````python
datasets = {}
for sigma in [0.5, 1.0, 2.0, 4.0, 100]:
    samples = np.random.normal(0, sigma, 8000)
    datasets[f"σ = {sigma}"] = (np.arange(len(samples)), samples)

isweep(datasets, title="KDE: Varying Spread", xlabel="Value", ylabel="Density", kind="bell")
````

### 2-Slider Sweep

Nested dict with 2 levels to 2 sliders appear automatically. For multivariable parameters to vary.

````python
x = np.linspace(0, 10, 500)

data = {}
for amp in [0.5, 1.0, 2.0, 5.0]:
    inner = {}
    for decay in [0.1, 0.3, 0.5, 1.0]:
        inner[f"decay = {decay}"] = (x, amp * np.exp(-decay * x) * np.sin(4 * x))
    data[f"amp = {amp}"] = inner

isweep(data, title="2D Sweep: Amplitude × Decay", xlabel="x", ylabel="y", kind="line")
````

### 3-Slider Sweep

Three levels of nesting to three sliders. Code should(?) work to N variables deep?

````python
x = np.linspace(0, 6, 400)

data = {}
for a in [1, 3, 5]:
    lvl2 = {}
    for b in [0.5, 1.0, 2.0]:
        lvl3 = {}
        for c in [1, 2, 4]:
            y = a * np.sin(c * x) * np.exp(-b * x)
            lvl3[f"c = {c}"] = (x, y)
        lvl2[f"b = {b}"] = lvl3
    data[f"a = {a}"] = lvl2

isweep(data, title="3D Sweep: a × b × c", xlabel="x", ylabel="y", kind="line")
````

## Slider + Overlay (`isweep_overlay`)

Slider picks a parameter group. Legend toggles traces. Legend state persists across slider moves.

````python
x = np.linspace(0, 10, 500)

groups = {}
for phase in [0, 45, 90, 135, 180]:
    phi = np.radians(phase)
    groups[f"phase = {phase}°"] = {
        "sin": (x, np.sin(x + phi)),
        "cos": (x, np.cos(x + phi)),
        "sum": (x, np.sin(x + phi) + np.cos(x + phi)),
    }

isweep_overlay(groups, title="Phase Shift Sweep", xlabel="x", ylabel="y", kind="line")
````

### 2-Slider + Overlay Example

Two nesting levels above the trace dict → two sliders + legend.

````python
x = np.linspace(0, 8, 400)

data = {}
for a in [1, 2, 5]:
    inner = {}
    for b in [0.2, 0.5, 1.0, 2.0]:
        traces = {
            "envelope":  (x, a * np.exp(-b * x)),
            "oscillation": (x, a * np.exp(-b * x) * np.cos(6 * x)),
        }
        inner[f"b = {b}"] = traces
    data[f"a = {a}"] = inner

isweep_overlay(data, title="2D Sweep + Overlay", xlabel="x", ylabel="y", kind="line")
````

## API Reference

```python
# Single trace
iplot(x, y, title=, xlabel=, ylabel=, kind=, color=, x_log=, y_log=)

# Overlay — click legend to hide/show
ioverlay({label: (x, y), ...}, title=, kind=, x_log=, y_log=)

# N-dimensional sweep — one slider per nesting level
isweep(nested_dict, title=, kind=, x_log=, y_log=)
#   1 level:  {label: (x, y)}
#   2 levels: {label: {label: (x, y)}}
#   N levels: keep nesting ...

# N-dimensional sweep + overlay — sliders + legend
isweep_overlay(nested_dict, title=, kind=, x_log=, y_log=)
#   innermost dict is the trace-overlay level: {trace: (x, y)}
#   every dict level above it becomes a slider
```
